```python
# filling the waterfront data as it is too crucial for the project to drop it
total_available = df['waterfront'].value_counts().sum()
yes_p = int(df['waterfront'].value_counts()['YES'] * 100) / total_available
no_p = int(df['waterfront'].value_counts()['NO'] * 100 ) / total_available
missing = df['waterfront'].isna().sum()
fill_no = int(no_p * missing/100)
fill_yes =int( missing - fill_no)
df['waterfront'].fillna('NO',limit=fill_no,inplace=True)
df['waterfront'].fillna('YES',limit =fill_yes,inplace=True)
df['waterfront'].value_counts()
```

```python
df['view'].value_counts()
```

```python
# filling the 'view' column
df['view'].fillna('NONE',inplace=True)
df['view'].value_counts()
```

# Column Names and Descriptions for King County Data Set
* `id` - Unique identifier for a house
* `date` - Date house was sold
* `price` - Sale price (prediction target)
* `bedrooms` - Number of bedrooms
* `bathrooms` - Number of bathrooms
* `sqft_living` - Square footage of living space in the home
* `sqft_lot` - Square footage of the lot
* `floors` - Number of floors (levels) in house
* `waterfront` - Whether the house is on a waterfront
  * Includes Duwamish, Elliott Bay, Puget Sound, Lake Union, Ship Canal, Lake Washington, Lake Sammamish, other lake, and river/slough waterfronts
* `view` - Quality of view from house
  * Includes views of Mt. Rainier, Olympics, Cascades, Territorial, Seattle Skyline, Puget Sound, Lake Washington, Lake Sammamish, small lake / river / creek, and other
* `condition` - How good the overall condition of the house is. Related to maintenance of house.
  * See the [King County Assessor Website](https://info.kingcounty.gov/assessor/esales/Glossary.aspx?type=r) for further explanation of each condition code
* `grade` - Overall grade of the house. Related to the construction and design of the house.
  * See the [King County Assessor Website](https://info.kingcounty.gov/assessor/esales/Glossary.aspx?type=r) for further explanation of each building grade code
* `sqft_above` - Square footage of house apart from basement
* `sqft_basement` - Square footage of the basement
* `yr_built` - Year when house was built
* `yr_renovated` - Year when house was renovated
* `zipcode` - ZIP Code used by the United States Postal Service
* `lat` - Latitude coordinate
* `long` - Longitude coordinate
* `sqft_living15` - The square footage of interior housing living space for the nearest 15 neighbors
* `sqft_lot15` - The square footage of the land lots of the nearest 15 neighbors

`sqft_basement` is actually float data type than why is it in object data type in our dataframe
> ## **Investigating 🔍**

```python
df['sqft_basement'].sort_values().unique()
```

> **'?'** is the problem

```python
df[(df.sqft_basement == '?')]
```

```python
# replacing the '?' with 0 as it is the most common and converting it into float
df['sqft_basement'] = df['sqft_basement'].replace('?',"0").astype('float')
df['sqft_basement'].sort_values().unique()
```

#### Identifying Continuous,Discrete and Categorical data 

```python
 # ploting each column against 'price'
for col in df.columns:
    fig,ax=plt.subplots()
    ax.scatter(df[col],df['price'],label=col)
    ax.legend()
```

```python
# these are the discrete_numeric variables from the scatter plots 
Discrete_numeric = ['bedrooms','bathrooms','floors','zipcode']
# these are the continuous variables from the scatterplots
Continuous = ['sqft_living','sqft_lot','sqft_above',
              'sqft_basement','lat','long','sqft_living15',
              'sqft_lot15','yr_built']

df_c = df.copy()
df_c.drop(columns=[x for x in Discrete_numeric],inplace=True)
```

```python
df_c.drop(columns=[x for x in Continuous],inplace=True)
```

```python
df_c.drop('price',inplace=True,axis=1)
```

```python
from datetime import datetime
# converting date from object dtype to datetime 
df_c['date'] = df_c['date'].apply(lambda x: datetime.strptime(x, '%m/%d/%Y'))

df_c['numeric_year'] = df_c['date'].apply(lambda x: x.year)
df_c['numeric_month'] = df_c['date'].apply(lambda x: x.month)
df_c['numeric_day'] = df_c['date'].apply(lambda x: x.day)
```

```python
X_continuous = df[[x for x in Continuous]]
```

```python
X_discrete = df[[x for x in Discrete_numeric]]
```

```python
df_c.info()
```

```python
X_discrete['numeric_year'] = df_c['numeric_year']
X_continuous['numeric_month'] = df_c['numeric_month']
X_continuous['numeric_day'] = df_c['numeric_day']
```

```python
X_continuous
```

```python
df_c.drop(['date','numeric_year','numeric_month','numeric_day'],axis=1)
```

```python
# these are the Categorical data
X_categorical= df_c
```

```python
y = df['price']
```

## Continuous variables

```python
X_continuous
```

```python
model = sm.OLS(y,sm.add_constant(X_continuous))
results = model.fit()
```

```python
results.summary()
```

## **Interpreting the Model**

**R-squared**: R-squared from the model Suggests about 61% of the variance in the dependent variable(price) is explained by the model

**Prob (F-statistic)**: This value is very close to zero, indicating that the overall regression model is highly significant.

**Const**: The constant(intercept) is approximately −60.7 million. This represents the predicted price when all other predictors are zero

> `sqft_living`: For each unit increase in square footage , the price of the house is expected to increase by $170.53.

> `sqft_lot`: Holding other variables constant, for each additional square foot of lot size, the price is expected to increase by $0.2275.

> `sqft_above`: An increase of one square foot  results in an increase in price by $91.15, assuming other variables are constant.

> `sqft_basement`: Each additional square foot in the basement is associated with an increase in price by $45.59, assuming other variables are constant.

> `lat`: For each unit increase in latitude, the price of the house is expected to increase by $621,900.

> `long`: Longitude also plays a role. For each unit increase in longitude , the price decreases by $296,300, assuming other variables are constant.

> `sqft_living15`: For each square foot increase , the price increases by $93.54, assuming other variables are constant.

> `sqft_lot15`: Each additional square foot decreases the price by $0.3791, assuming other variables are constant.

> `yr_built`: For each year increase in the year of construction, the price decreases by $1,710.74, assuming other variables are constant.

> `numeric_month`:  as the numeric month increases, the price decreases by $2,792.96, assuming other variables are constant.

> `numeric_day`:  as the numeric day increases, the price decreases by $359.74, assuming other variables are constant.
